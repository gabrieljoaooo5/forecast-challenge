## 2.2: Models

### Baseline model

```python
import sklearn.metrics as metrics

dataframe = pd.concat([df.shift(1), df], axis=1)
dataframe.columns = ['t', 't+1']

# split into train and test sets
X = dataframe.values
train2, test2 = X[1:1310], X[1310:]
train_X, train_y = train2[:,0], train2[:,1]
test_X, test_y = test2[:,0], test2[:,1]

# persistence model
def model_persistence(x):
    return x

# walk-forward validation
predictions = list()
for x in test_X:
    yhat = model_persistence(x)
    predictions.append(yhat)
rmse = np.sqrt(metrics.mean_squared_error(test_y, predictions))
print('Test RMSE: %.3f' % rmse)
```

### Autoregression model

```python
from statsmodels.tsa.ar_model import AutoReg

def autoRegModel(df, lags, train_size=1310):
    # split dataset
    X = df.values
    train, test = X[1:train_size], X[train_size:]
    # train autoregression
    model = AutoReg(train, lags=lags)
    model_fit = model.fit()
    coef = model_fit.params
    # walk forward over time steps in test
    history = train[len(train)-lags:]
    history = [history[i] for i in range(len(history))]
    predictions = list()
    for t in range(len(test)):
        length = len(history)
        lag = [history[i] for i in range(length-lags,length)]
        yhat = coef[0]
        for d in range(lags):
            yhat += coef[d+1] * lag[lags-d-1]
        obs = test[t]
        predictions.append(yhat)
        history.append(obs)
    rmse = np.sqrt(metrics.mean_squared_error(test, predictions))
    return rmse, predictions
def evaluate_AR_models(df, lags):
    best_score, best_cfg = float("inf"), None
    for lag in lags:
        try:
            rmse = autoRegModel(df, lag)[0]
            if rmse < best_score:
                best_score, best_cfg = rmse, lag
            print('AR%s RMSE=%.3f' % (lag,rmse))
        except:
            continue
    print('Best AR%s RMSE=%.3f' % (best_cfg, best_score))
    
lags = [1, 7, 14, 21, 28, 35, 42, 49, 56, 63, 70]
evaluate_AR_models(df, lags)
```

### ARIMA model with Grid Search

```python
from statsmodels.tsa.arima_model import ARIMA

def evaluate_arima_model(X, arima_order):
    # prepare training dataset
    train_size = 1310
    train, test = X[0:train_size], X[train_size:]
    history = [x for x in train]
    # make predictions
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit(disp=0)
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(test[t])
    rmse = np.sqrt(metrics.mean_squared_error(test, predictions))
    return rmse

def evaluate_ARIMA_models(df, p_values, d_values, q_values):
    df = df.astype('float32')
    best_score, best_cfg = float("inf"), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    rmse = evaluate_arima_model(df, order)
                    if rmse < best_score:
                        best_score, best_cfg = rmse, order
                    print('ARIMA%s RMSE=%.3f' % (order,rmse))
                except:
                    continue
    print('Best ARIMA%s RMSE=%.3f' % (best_cfg, best_score))
    
p_values = range(0, 3)
d_values = range(0, 3)
q_values = range(0, 3)
evaluate_ARIMA_models(df.values, p_values, d_values, q_values)
```

### Aproach 1

```python
from dateutil.parser import parse

def isDate(string, fuzzy=False):
    try: 
        parse(string, fuzzy=fuzzy)
        return True
    except ValueError:
        return False
    
def setShiftByTime(h):
    if (isDate(h) == False):
        return 'remove'
    h = int(h.split(':')[0])
    if (h >= 5 and h < 12):
        return 'manha'
    if (h >= 12 and h < 17):
        return 'tarde'
    if (h >= 17 and h < 22):
        return 'noite'
    else:
        return 'madrugada'
    
def filterDf(df, columns):
    return df[columns].dropna()

def getOccurrencesPerDay(df):
    return pd.DataFrame(df.data.value_counts())

def filterDfByShift(df, shift):
    df = getOccurrencesPerDay(df[df.turno == shift])    
    df.sort_index(inplace=True)
    df.index.name = 'Date'
    df.columns=  ['Total_Accidents']
    return df

def fillWithZero(df, date):
    if date in df.index:
        return df[df.index == date].Total_Accidents.iloc[0]
    return 0
```

```python
df_2015 = pd.read_csv("../dataset/acidentes-transito-2015.csv", parse_dates=['data'])
df_2016 = pd.read_csv("../dataset/acidentes_2016.csv", parse_dates=['data'])
df_2017 = pd.read_csv("../dataset/acidentes_2017.csv", parse_dates=['data'])
df_2018 = pd.read_csv("../dataset/acidentes_2018.csv", parse_dates=['data'])
df_2019 = pd.read_csv("../dataset/acidentes-2019.csv", parse_dates=['data'])
```

```python
df_time_2015 = filterDf(df_2015, ['data', 'hora'])
df_time_2016 = filterDf(df_2016, ['data', 'hora'])
df_time_2017 = filterDf(df_2017, ['data', 'hora'])
df_time_2018 = filterDf(df_2018, ['data', 'hora'])
df_time_2019 = filterDf(df_2019, ['data', 'hora'])
```

```python
frames = [df_time_2015, df_time_2016, df_time_2017, df_time_2018, df_time_2019]

shift_df = pd.concat(frames)
shift_df['turno'] = shift_df.hora.apply(lambda h: setShiftByTime(h))
shift_df = shift_df[shift_df.turno != 'remove']
shift_df = shift_df[['data', 'turno']]

morning_df = filterDfByShift(shift_df, 'manha')
afternoon_df = filterDfByShift(shift_df, 'tarde')
night_df = filterDfByShift(shift_df, 'noite')
dawn_df = filterDfByShift(shift_df, 'madrugada')

shift_df = df.reset_index()
shift_df['manha'] = shift_df.Date.apply(lambda d: fillWithZero(morning_df, d))
shift_df['tarde'] = shift_df.Date.apply(lambda d: fillWithZero(afternoon_df, d))
shift_df['noite'] = shift_df.Date.apply(lambda d: fillWithZero(night_df, d))
shift_df['madrugada'] = shift_df.Date.apply(lambda d: fillWithZero(dawn_df, d))
```

### Multivariate data 
* Since multivariate data is often more difficult to work with, I will not train the model. Instead, I will show a technique of extracting the neighborhood and the accident zone. This can be useful for future work with this case.

```python
df_neighborhood = pd.read_csv("../dataset/bairros_recife.csv")
df_neighborhood.Bairros = df_neighborhood.Bairros.apply(lambda b: b.split(', '))
df_neighborhood
```

```python
import unidecode

def remveNullAndAccents(df):
    df = df[df.bairro.isnull() == False]
    df.bairro = df.bairro.apply(lambda b: unidecode.unidecode(b.lower()))
    return df

df_neighborhood_2015 = remveNullAndAccents(df_2015)
df_neighborhood_2016 = remveNullAndAccents(df_2016)
df_neighborhood_2017 = remveNullAndAccents(df_2017)
df_neighborhood_2018 = remveNullAndAccents(df_2018)
df_neighborhood_2019 = remveNullAndAccents(df_2019)
```

```python
def searchRPA(neighborhood):
    for count in range (0, 6):
        if neighborhood in df_neighborhood.iloc[count].Bairros:
            return df_neighborhood.iloc[count].RPA
        

        
df_neighborhood_2015['RPA'] = df_neighborhood_2015.bairro.apply(lambda n: searchRPA(n))
df_neighborhood_2015 = filterDf(df_neighborhood_2015, ['data', 'hora', 'bairro', 'RPA'])

df_neighborhood_2016['RPA'] = df_neighborhood_2016.bairro.apply(lambda n: searchRPA(n))
df_neighborhood_2016 = filterDf(df_neighborhood_2016, ['data', 'hora', 'bairro', 'RPA'])

df_neighborhood_2017['RPA'] = df_neighborhood_2017.bairro.apply(lambda n: searchRPA(n))
df_neighborhood_2017 = filterDf(df_neighborhood_2017, ['data', 'hora', 'bairro', 'RPA'])

df_neighborhood_2018['RPA'] = df_neighborhood_2018.bairro.apply(lambda n: searchRPA(n))
df_neighborhood_2018 = filterDf(df_neighborhood_2018, ['data', 'hora', 'bairro', 'RPA'])

df_neighborhood_2019['RPA'] = df_neighborhood_2019.bairro.apply(lambda n: searchRPA(n))
df_neighborhood_2019 = filterDf(df_neighborhood_2019, ['data', 'hora', 'bairro', 'RPA'])
```
